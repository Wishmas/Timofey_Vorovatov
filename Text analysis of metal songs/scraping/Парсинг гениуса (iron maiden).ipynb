```python
import requests
from bs4 import BeautifulSoup
import lxml
from time import sleep
import re
```

```python
alb_list = ['Iron Maiden','Killers','The Number Of The Beast','Piece Of Mind','Powerslave','Somewhere In Time',
            'Seventh Son Of A Seventh Son','No Prayer For The Dying','Fear Of The Dark','The X Factor','Brave New World']
songs_list = []
```

```python
headers =  {"User-Agent": "Mozilla/4.0 (Windows NT 7.0; Win64; x64) AppleWebKit/987.32 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/535.38"}
lines = []
for alb in alb_list:
    sleep(2)
    al = alb
    al_rest = al[1:].lower().replace("’",' ').split(' ')
    al_fin = al[0] + '-'.join(al_rest)
    url = f'https://genius.com/albums/Iron-maiden/{al_fin}'
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, 'lxml')
    tbox = soup.find_all('h3',class_="chart_row-content-title") 
    for i in tbox:
        it = i.text.replace('\n','').replace('Lyrics','').strip().lower().split(' ')
        it = '-'.join(it)
        it = it.replace(',','').replace('(','').replace(')','').replace('’','')
        songs_list.append(it)
print('done')

```

```python
headers =  {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/789.36 (KHTML, like Gecko) Chrome/101.0.0.0 Safari/532.36"}
lines = []
for n in songs_list:
    try:
        sleep(1)
        url = f'https://genius.com/Iron-maiden-{n}-lyrics'
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.text, 'lxml')
        tbox = soup.find_all('div',class_="Lyrics__Container-sc-1ynbvzw-6 YYrds") 
        for t in tbox:
            for i in t.contents:
                it = i.text
                if it != '' and ('[' not in it or ']' not in it): 
                    result = it[0]
                    for letter in it[1:]:
                        if letter.isupper():
                             result += f' {letter}'
                        else:
                            result += letter
                    lines.append(result) 
    except:
        pass
    print(n)
print('list done')
with open(r"iron_maiden_texts.txt", "w",encoding="utf-8") as file:
    for  line in lines:
        file.write(line + '\n')  
print('File done')
```
